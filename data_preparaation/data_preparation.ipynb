```python
import pandas as pd
import os
```

```python
source_file_path = os.path.relpath(r"../data/PyCode_Vul%20test-set.csv")
df = pd.read_csv(source_file_path)

df["index"] = df.index
```

```python
storage_path = os.path.abspath(r"../data/py_test_data_small")
os.makedirs(storage_path, exist_ok=True)
```

```python
code_col = "function_code"
name_col = "repo"
index_id = "index"
```

```python
try:
    df = df.sample(300)
    counter = 1
    for index, row in df.iterrows():
            raw_name = str(row[name_col]).strip()
            code_content = row[code_col]
            index_num = row[index_id]

            if pd.isna(code_content):
                print(f"Skipping row {index}: No code found.")
                continue

            safe_name = raw_name.replace(" ", "_")
            safe_name = safe_name.replace("/","_")

            file_name = f"{safe_name}_{index_num}.py"
            file_path = os.path.join(storage_path, file_name)

            with open(file_path, "w+", encoding="utf-8") as py_file:
                py_file.write(str(code_content))

            print(f"Saved: {file_path}")

            counter += 1
except KeyError as e:
    print(
        f"Error: Column {e} not found in the CSV. Please check your column names."
    )
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
